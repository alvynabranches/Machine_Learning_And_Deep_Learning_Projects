```python
audio, sr = librosa.load(os.path.join(path, file))
```

```python
%%time
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(32,20))
ax[0, 0].set_title('Audio')
ax[0, 0].plot(audio)
ax[0, 1].set_title('Chroma Cens')
ax[0, 1].plot(librosa.feature.chroma_cens(audio, sr))
ax[0, 2].set_title('Chroma CQT')
ax[0, 2].plot(librosa.feature.chroma_cqt(audio, sr))
ax[0, 3].set_title('Chroma STFT')
ax[0, 3].plot(librosa.feature.chroma_stft(audio, sr))
ax[0, 4].set_title('Delta')
# ax[0, 4].plot(librosa.feature.delta(audio))
ax[1, 0].set_title('Fourier Tempogram')
ax[1, 0].plot(librosa.feature.fourier_tempogram(audio, sr))
ax[1, 1].set_title('Melspectrogram')
ax[1, 1].plot(librosa.feature.melspectrogram(audio, sr))
ax[1, 2].set_title('MFCC')
ax[1, 2].plot(librosa.feature.mfcc(audio, sr))
ax[1, 3].set_title('Poly Features - Order 1')
ax[1, 3].plot(librosa.feature.poly_features(S=np.abs(librosa.stft(audio)), order=1))
ax[1, 4].set_title('Auto Correlate')
ax[1, 4].plot(librosa.feature.rhythm.autocorrelate(audio))
ax[2, 0].set_title('STFT')
ax[2, 0].plot(librosa.feature.rhythm.stft(audio))
ax[2, 1].set_title('RMS')
ax[2, 1].plot(librosa.feature.rms(audio))
ax[2, 2].set_title('CQT')
ax[2, 2].plot(librosa.feature.spectral.cqt(audio, sr))
ax[2, 3].set_title('Estimate Tunning')
ax[2, 3].plot(librosa.feature.spectral.estimate_tuning(audio, sr))
ax[2, 4].set_title('Hybrid CQT')
ax[2, 4].plot(librosa.feature.spectral.hybrid_cqt(audio, sr))
ax[3, 0].set_title('Spectral Bandwidth')
ax[3, 0].plot(librosa.feature.spectral_bandwidth(audio, sr))
ax[3, 1].set_title('Spectral Centroid')
ax[3, 1].plot(librosa.feature.spectral_centroid(audio, sr))
ax[3, 2].set_title('Spectral Contrast')
ax[3, 2].plot(librosa.feature.spectral_contrast(audio, sr))
ax[3, 3].set_title('Spectral Flatness')
ax[3, 3].plot(librosa.feature.spectral_flatness(audio))
ax[3, 4].set_title('Spectral Rolloff')
ax[3, 4].plot(librosa.feature.spectral_rolloff(audio, sr))
ax[4, 0].set_title('Tempogram')
ax[4, 0].plot(librosa.feature.tempogram(audio, sr))
ax[4, 1].set_title('Tonnetz')
ax[4, 1].plot(librosa.feature.tonnetz(audio, sr))
ax[4, 2].set_title('Zero Crossing Rate')
ax[4, 2].plot(librosa.feature.zero_crossing_rate(audio))
ax[4, 3].set_title('Poly Features - Order 7')
ax[4, 3].plot(librosa.feature.poly_features(S=np.abs(librosa.stft(audio)), order=7))
ax[4, 4].set_title('Spectrogram')
librosa.display.specshow(librosa.amplitude_to_db(np.abs(librosa.stft(audio)), ref=np.max), y_axis='log', x_axis='time', ax=ax[4, 4])
plt.savefig(f'graphs/{file.split(".")[0]}.jpg', dpi=1280)
```
